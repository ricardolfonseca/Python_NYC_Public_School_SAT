SAT scores are significant for students, colleges, and various stakeholders involved in education.

### First question: What are the top 10 schools with the highest average math scores?

This exercise focuses on identifying the top 10 schools based on their average math scores.

### Solution Summary

The following steps were taken to answer this question:

1. **Load Libraries and Data**:
   - Import the necessary libraries: `pandas` for data manipulation and `matplotlib.pyplot` for data visualization.
   - Load the data from `schools.csv` into a `pandas` DataFrame called `schools`.

2. **Preview the Data**:
   - Use `.head()` to check the structure of the data and inspect the columns.

3. **Filter Schools by Math Score**:
   - Identify schools where the `average_math` score is at least 80% of the maximum possible score (800). This threshold (640) isolates schools with stronger math performance, resulting in a subset stored in `schools_above_80_math`.

4. **Select Relevant Columns**:
   - Retain only the `school_name` and `average_math` columns to simplify the data. This reduced dataset is stored in `schools_above_80_math_drop_columns`.

5. **Sort Schools by Average Math Score**:
   - Sort the filtered schools by `average_math` in descending order to rank schools from highest to lowest average math scores. The sorted DataFrame is named `best_math_schools`.

6. **Display Top 10 Schools**:
   - Print the top 10 schools with the highest average math scores, setting `school_name` as the index for a clearer output.

7. **Visualize Results with a Bar Plot**:
   - Generate a bar plot to visualize the top 10 schools based on average math scores.
   - Customize the plot by adding a title and adjusting the font size for readability.

### Output

The code outputs:
- A list of the top 10 schools with the highest average math scores.
- A bar plot that visually represents the top 10 schoolsâ€™ performance in math.

### Prerequisites

Ensure that:
- `schools.csv` is in the correct directory, as well as the code file.
- `pandas` and `matplotlib` libraries are installed.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data
schools = pd.read_csv("../Dataset/schools.csv")

# Preview the data
schools.head()

#Get schools with 'average_math' equal to or higher than 80% of 800
schools_above_80_math = schools[schools['average_math'] >= (800*0.8)]

#Select all rows, only 'school_name' and 'average_math' columns
schools_above_80_math_drop_columns = schools_above_80_math.loc[:,('school_name', 'average_math')]

#Sort values descending
best_math_schools = schools_above_80_math_drop_columns.sort_values('average_math', ascending=False)

#Print with index 'school_name'
print('The top 10 schools with the highest average math scores are:')
print(best_math_schools.set_index('school_name'))

#Create and customize bar plot
best_math_schools.set_index('school_name').plot(kind="bar")
plt.title('Top 10 schools with the highest average math scores')
plt.legend('')
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

### Second question: What are the top 10 schools with the highest SAT scores?

This exercise aims to identify the top 10 schools with the highest combined SAT scores in math, reading, and writing.

### Solution Summary

The following steps were taken to answer the question:

1. **Load Libraries and Data**:
   - Import `pandas` for data manipulation and `matplotlib.pyplot` for visualization.
   - Load data from `schools.csv` into a `pandas` DataFrame named `schools`.

2. **Calculate Total SAT Score**:
   - Define the columns to sum: `average_math`, `average_reading`, and `average_writing`.
   - Add a new column, `total_SAT`, to the DataFrame that represents the combined SAT score for each school by summing the selected columns.

3. **Sort Schools by Total SAT**:
   - Sort the schools by the `total_SAT` column in descending order to rank schools from highest to lowest combined SAT scores.

4. **Select Top 10 Schools**:
   - Extract the top 10 rows from the sorted DataFrame, retaining only the `school_name` and `total_SAT` columns. This subset is stored in `top_10_schools`.
   - Display the top 10 schools and their total SAT scores, using `school_name` as the index for readability.

5. **Calculate and Plot Average SAT Score**:
   - Calculate the average `total_SAT` for the top 10 schools.
   - Create a bar plot of the top 10 schools' total SAT scores, with customized labels and title for clarity.
   - Add a horizontal line representing the average SAT score of the top 10 schools, colored in red with a dashed line style for emphasis.

6. **Display the Plot**:
   - Include a legend indicating the average SAT score line, then show the plot.

### Output

The code outputs:
- A list of the top 10 schools with the highest combined SAT scores.
- A bar plot that visually represents the top 10 schools' SAT scores with an indication of the average score.

### Prerequisites

Ensure that:
- `schools.csv` is in the correct directory, as well as the code file.
- `pandas` and `matplotlib` libraries are installed.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data
schools = pd.read_csv("../Dataset/schools.csv")

# Assign columns to SUM
cols = ['average_math', 'average_reading', 'average_writing']

# Add column total_SAT to dataframe, sort descending
schools['total_SAT'] = schools[cols].sum(axis=1)
schools = schools.sort_values('total_SAT', ascending=False)

# Slice by top 10 rows and only selected columns
top_10_schools = schools.iloc[:10][['school_name', 'total_SAT']]
print('The top 10 schools with the highest SAT scores are:')
print(top_10_schools.set_index('school_name'))

# Calculate average SAT for top 10 schools
average_SAT_top_10 = top_10_schools['total_SAT'].mean()

# Create and customize bar plot
ax = top_10_schools.set_index('school_name').plot(kind="bar")
plt.title('Top 10 schools with the highest SAT scores')
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(fontsize=8)
plt.xlabel('')
plt.ylabel('Total SAT Score')
plt.legend('')

# Add a horizontal line for the average SAT score of the top 10 schools
plt.axhline(y=average_SAT_top_10, color='r', linestyle='--', label=f'Average SAT (Top 10): {average_SAT_top_10:.2f}')

# Display the legend
plt.legend(['Average SAT (Top 10)'])

# Show plot
plt.show()
```

### Third question: What is the borough with the schools that have the largest standard deviation for total SAT scores, and the average SAT scores?

This analysis determines borough with the schools that exhibit the largest variation in SAT scores and identifying the average SAT score within that borough.

### Solution Summary

The following steps were taken to answer this question:

1. **Load Libraries and Data**:
   - Import `pandas` for data manipulation.
   - Load the data from `schools.csv` into a `pandas` DataFrame called `schools`.

2. **Preview the Data**:
   - Use `.head()` to examine the structure of the data and confirm column names and data types.

3. **Calculate Total SAT Score**:
   - Define the columns to sum: `average_math`, `average_reading`, and `average_writing`.
   - Add a new column, `total_SAT`, to represent the combined SAT score for each school by summing these columns.

4. **Group Data by Borough**:
   - Group the data by `borough` and calculate the following:
     - **Number of Schools**: Count the number of schools in each borough.
     - **Average SAT**: Compute the mean of `total_SAT` scores in each borough, rounded to two decimal places.
     - **Standard Deviation of SAT**: Compute the standard deviation of `total_SAT` scores in each borough, rounded to two decimal places.

5. **Identify Borough with Largest SAT Standard Deviation**:
   - Sort the grouped data by `std_SAT` in descending order.
   - Select the borough with the largest standard deviation in SAT scores, stored in `largest_std_dev`.
   - Display the name of this borough along with its average SAT score and standard deviation.

### Output

The code outputs:
- The borough with the schools that have the largest standard deviation for SAT scores.
- The average SAT score and the standard deviation of SAT scores in that borough.

### Prerequisites

Ensure that:
- `schools.csv` is in the correct directory, as well as the code file.
- `pandas` library is installed.

```python
import pandas as pd

# Read in the data
schools = pd.read_csv("../Dataset/schools.csv")

# Preview the data
schools.head()

# Assign columns to SUM
cols = ['average_math', 'average_reading', 'average_writing']

# Add column total_SAT to dataframe
schools['total_SAT'] = schools[cols].sum(axis=1)

# Group by 'borough' and calculate the desired statistics
schools_grouped_borough = (
    schools.groupby('borough')
    .agg(num_schools=('school_name', 'count'))
    .assign(
        average_SAT=schools.groupby('borough')['total_SAT'].mean().round(2),
        std_SAT=schools.groupby('borough')['total_SAT'].std().round(2)
    )
)

# Sort by 'std_SAT' in descending order and show first row
schools_grouped_borough_sorted = schools_grouped_borough.sort_values('std_SAT', ascending=False)
largest_std_dev = schools_grouped_borough_sorted.iloc[[0]]
print(f'The borough with the schools that have the largest standard deviation for total SAT scores is:')
print(f'{largest_std_dev}')
```
